## 7. Missing Value Analysis

```python
missing = pd.DataFrame({
    "Missing Count": df.isnull().sum(),
    "Missing Percentage": (df.isnull().sum() / len(df)) * 100
})

missing = missing.sort_values(
    by="Missing Percentage",
    ascending=False
)

missing.head(20)
```

```python
missing.describe()
```

```python
empty_columns = missing[
    missing["Missing Percentage"] == 100
]

print(f"Completely Empty Columns : {len(empty_columns)}")
```

```python
empty_columns.head(20)
```

# 11. Audit Conclusions

## Dataset Summary

- Total Records: 9,082
- Total Features: 3,925
- Target Column: F3924

## Data Quality Findings

- No duplicate records.
- 63 columns contain 100% missing values.
- Average missingness is high.
- Missing values are unevenly distributed across features.
- Dataset contains 8 categorical features.

## Target Variable

- Fraud Class (1): 81 samples (0.89%)
- Normal Class (0): 9001 samples (99.11%)

The dataset is highly imbalanced.

Accuracy is not an appropriate evaluation metric.

Future models will be evaluated using:

- Precision
- Recall
- F1-score
- ROC-AUC
- PR-AUC

## Preprocessing Decisions

- Remove index column.
- Remove fully empty columns.
- Encode categorical variables.
